# Cleaning Data

```python
df.dropna(inplace=True)
useless_columns = ['user_id', 'user_name', 'review_id', 'review_title','review_content', 'img_link', 'about_product', 'product_link', 'product_id']
df.drop(useless_columns, axis=1, inplace=True)
```

### Changing Data Format

```python
def format_money(x):
    formatted = x[1:].replace(',', '')
    formatted = np.multiply(np.float32(formatted), 0.0064)
    return np.round(formatted, 2)
    

df[['discounted_price', 'actual_price']] = df[['discounted_price', 'actual_price']].map(format_money)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(np.uint8)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce').astype(np.float32)
df.dropna(subset=['rating'], inplace=True)
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(np.uint16)
```

### Adding Columns

```python
categories = df.join(df['category'].str.split('|', expand=True))

df['main_category'] = categories[0]
df['sub_category'] = categories[1]

df.drop('category', axis=1, inplace=True)


C = df['rating'].mean()
m = 100
df['weighted_rating'] = (df['rating'] * df['rating_count']) / (df['rating_count'] + m) + (C * m) / (df['rating_count'] + m)


conditions = [
    df['weighted_rating'] > 4.5,
    df['weighted_rating'] > 3.5,
    df['weighted_rating'] <= 3.5
]

labels = ['Good Rated', 'Average Rated', 'Poor Rated']

df['rating_label'] = np.select(conditions, labels)
```

### Checking Duplicates

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

# Correlation

```python
corr_df = df[['actual_price', 'discount_percentage', 'discounted_price', 'rating', 'rating_count', 'weighted_rating']]
corr = corr_df.corr()

sns.heatmap(corr, annot=True)
sns.pairplot(corr_df)

plt.show()
```

# Analysing

### Counting All Groups of Rated Products

```python
sns.countplot(x='rating_label', hue='rating_label', data=df)
plt.title('Product Rating Distribution')
plt.show()
```

__Explanation__: Most products have an average rating.

### Poor and Good Rated Products

```python
good_rated = df[df['rating_label'] == 'Good Rated'][['product_name', 'weighted_rating']].sort_values(by='weighted_rating', ascending=False)
bad_rated = df[df['rating_label'] == 'Poor Rated'][['product_name', 'weighted_rating']].sort_values(by='weighted_rating', ascending=True)

fig, axs = plt.subplots(2, 1, figsize=(15, 25))

ax1 = sns.barplot(ax=axs[0], data=bad_rated, x='weighted_rating', y='product_name', hue='product_name')
ax1.set_title('Bad Rated Product')
for i in ax1.containers:
    ax1.bar_label(i, fontsize=10)

ax2 = sns.barplot(ax=axs[1], data=good_rated, x='weighted_rating', y='product_name', hue='product_name')
ax2.set_title('Bad Rated Product')
for i in ax2.containers:
    ax2.bar_label(i, fontsize=10)

plt.show()
```

__Explanation__: The worst-rated product is "ROYAL STEP," and the best-rated product is "Swiffer Instant.

### Well-Rated Products with a High Discount Rate

```python
high_discount_rate = df[(df['discount_percentage'] >= 60) & (df['rating_label'] == 'Good Rated')][['product_name', 'discount_percentage', 'rating_count', 'weighted_rating']]
high_discount_rate.sort_values(by='weighted_rating', ascending=False)
```

__Explanation__: Maybe we can reduce the discount percentage for this product.

### Products with a Small Rating Count but a Good Rating and a Low Discount Rate

```python
low_discount_rate = df[(df['discount_percentage'] <= 30) & (df['rating'] >= 4) & (df['rating_count'] <= m)][['product_name', 'discount_percentage', 'rating', 'rating_count']]
low_discount_rate.sort_values(by='rating', ascending=False)
```

__Explanation__: These products have the potential for higher sales.

### Products Actual Price, Rating Count, Weighted Rating, Discount Rate  Distribution

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20,10))

sns.histplot(ax=axs[0,0], data=df, x='actual_price', kde=True, bins=50, color='r')
sns.histplot(ax=axs[0,1], data=df, x='rating_count', kde=True, bins=50, color='g')
sns.histplot(ax=axs[1,0], data=df, x='weighted_rating', kde=True, bins=20, color='b')
sns.histplot(ax=axs[1,1], data=df, x='discount_percentage', kde=True, bins=20, color='y')

plt.show()
```

### Checking Main Categories and Subcategories

```python
main_categories = df['main_category'].value_counts().reset_index()
sub_categories = df['sub_category'].value_counts().reset_index()

fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(12,13))

ax1 = sns.barplot(ax=axs[0], x='count', y='main_category', hue='main_category', data=main_categories)
axs[0].set(title='Main Categories')
for i in ax1.containers:
    ax1.bar_label(i, fontsize=10)

ax2 = sns.barplot(ax=axs[1], x='count', y='sub_category', hue='sub_category',data=sub_categories)
axs[1].set(title='Sub Categories')
for i in ax2.containers:
    ax2.bar_label(i, fontsize=10)

plt.show()
```

__Explanation__: Electronics, Home & Kitchen, and Computer & Accessories have the most products. Accessories & Peripherals covers the most products within Computer & Accessories and has the highest number of products across other subcategories. Kitchen & Home Appliances covers the most products within Home & Kitchen, while Home Theater, TV & Video, and Mobiles & Accessories cover the most products within Electronics.
