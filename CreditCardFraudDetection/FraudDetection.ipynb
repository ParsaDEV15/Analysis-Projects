```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, KFold
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_recall_curve, log_loss
```

# Getting Data

```python
data = pd.read_csv('creditcard.csv')
```

# Creating Model

```python
X = data.drop(['Class', 'Time', 'Amount'], axis=1)
y = data['Class']

ax = sns.countplot(x='Class', data=data)
ax.bar_label(ax.containers[0], fontsize=10)
plt.show()
```

__Explanation__: Class 0 has a significantly larger number of instances than Class 1, which leads to imbalance issues.

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=128)

smote = SMOTE(random_state=128)
train_resample_X, train_resample_y = smote.fit_resample(train_X, train_y)

logreg = LogisticRegression(solver='liblinear')
logreg.fit(train_resample_X, train_resample_y)

y_scores = logreg.predict_proba(test_X)[:, 1]
precisions, recalls, thresholds = precision_recall_curve(test_y, y_scores)

plt.figure(figsize=(8, 6))
plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.title("Precision-Recall vs Threshold")
plt.show()


hypo_threshold = 0.8
start_index = np.argmax(thresholds >= hypo_threshold)
max_index = len(thresholds)
target_thresholds = thresholds[start_index:max_index+1]

f1_scores = []
tested_thresholds = []

for threshold in target_thresholds:
    y_pred = (y_scores >= threshold).astype(int)
    score = f1_score(test_y, y_pred)
    f1_scores.append(score)
    tested_thresholds.append(threshold)

best_threshold = target_thresholds[np.argmax(f1_scores)]
y_pred_new = (y_scores >= best_threshold).astype(int)

kf = StratifiedKFold(n_splits=5, random_state=128, shuffle=True)
scores = cross_val_score(logreg, X, y, cv=kf, scoring='accuracy')

print(classification_report(test_y, y_pred_new))
print('Log Loss:', log_loss(test_y, y_pred_new))
print('Scores:', scores)
print('Accuracy Avg:', scores.mean())
```

```python
confusion_matrix = confusion_matrix(test_y, y_pred_new)
group_names = ['True Pos','False Neg','False Pos','True Neg']
group_count = [f'{value}' for value in confusion_matrix.flatten()]
labels = [f'{group_names[i]} \n {group_count[i]}' for i in range(4)]
labels = np.array(labels).reshape(2,2)
sns.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Blues')
plt.title("Confusion matrix")
plt.show()
```
